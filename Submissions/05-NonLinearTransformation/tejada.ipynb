```python
# import packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from scipy.stats import linregress
from sklearn.metrics import r2_score
```

# A. Nonlinear Transformation

```python
# loading data
cn_data = pd.read_csv("1_china_gdp.csv")
x, y = cn_data.iloc[:, 0], cn_data.iloc[:, 1]
x, y
```

### Model 0: Using Standard Linear Regression

```python
# Get slope and intercept
slope, intercept, rvalue, pvalue, stderr = linregress(x, y)

# Predict using the linear regression equation
y_pred = slope * x + intercept
```

```python
residual = y - y_pred

# Residual plot
plt.scatter(x, residual, c='red', label='Residuals')
plt.axhline(y=0, color='blue', linestyle='--', label='Zero Residual Line')
plt.xlabel('X')
plt.ylabel('Residuals')
plt.legend()
plt.title('Residual Plot Using Standard Linear Regression')
plt.grid(True)
plt.show()
```

```python
r2 = r2_score(y, y_pred)
print(f"The coefficient of determination or r squared is {r2}")
```

### Model 1: Using Exponential Model

```python
# Get slope and intercept
y_new = np.log(y)
slope, intercept, rvalue, pvalue, stderr = linregress(x, y_new)

# Predict using the modified equation that is suitable for nonlinear data
y_pred = np.power(10, intercept + slope * x)

# Compute r2 again
r2 = r2_score(y, y_pred)
print(f"The coefficient of determination or r squared is {r2}")
```

```python
residual = y - y_pred

# Residual plot
plt.scatter(x, residual, c='red', label='Residuals')
plt.axhline(y=0, color='blue', linestyle='--', label='Zero Residual Line')
plt.xlabel('X')
plt.ylabel('Residuals')
plt.legend()
plt.title('Residual Plot Using Exponential Model')
plt.grid(True)
plt.show()
```

### Model 2: Using Quadratic Model

```python
# Get slope and intercept
y_new = np.sqrt(y)
slope, intercept, rvalue, pvalue, stderr = linregress(x, y_new)

# Predict using the modified equation that is suitable for nonlinear data
y_pred = np.square(intercept + slope * x)

# Compute r2 again
r2 = r2_score(y, y_pred)
print(f"The coefficient of determination or r squared is {r2}")
```

```python
residual = y - y_pred

# Residual plot
plt.scatter(x, residual, c='red', label='Residuals')
plt.axhline(y=0, color='blue', linestyle='--', label='Zero Residual Line')
plt.xlabel('X')
plt.ylabel('Residuals')
plt.legend()
plt.title('Residual Plot Using Reciprocal Model')
plt.grid(True)
plt.show()
```

### Model 3: Using Reciprocal Model

```python
# Get slope and intercept
y_new = np.reciprocal(y)
slope, intercept, rvalue, pvalue, stderr = linregress(x, y_new)

# Predict using the modified equation that is suitable for nonlinear data
y_pred = np.reciprocal(intercept + slope * x)

# Compute r2 again
r2 = r2_score(y, y_pred)
print(f"The coefficient of determination or r squared is {r2}")
```

```python
residual = y - y_pred

# Residual plot
plt.scatter(x, residual, c='red', label='Residuals')
plt.axhline(y=0, color='blue', linestyle='--', label='Zero Residual Line')
plt.xlabel('X')
plt.ylabel('Residuals')
plt.legend()
plt.title('Residual Plot Using Reciprocal Model')
plt.grid(True)
plt.show()
```

### Model 4: Using Logarithmic Model

```python
# Get slope and intercept
x_new = np.log(x)
slope, intercept, rvalue, pvalue, stderr = linregress(x_new, y)

# Predict using the modified equation that is suitable for nonlinear data
y_pred = np.add(intercept, slope * np.log(x_new))

# Compute r2 again
r2 = r2_score(y, y_pred)
print(f"The coefficient of determination or r squared is {r2}")
```

```python
residual = y - y_pred

# Residual plot
plt.scatter(x, residual, c='red', label='Residuals')
plt.axhline(y=0, color='blue', linestyle='--', label='Zero Residual Line')
plt.xlabel('X')
plt.ylabel('Residuals')
plt.legend()
plt.title('Residual Plot Using Logarithmic Model')
plt.grid(True)
plt.show()
```

### Model 5: Using Power Model

```python
# Get slope and intercept
x_new = np.log(x)
y_new = np.log(y)
slope, intercept, rvalue, pvalue, stderr = linregress(x_new, y_new)

# Predict using the modified equation that is suitable for nonlinear data
y_pred = np.power(10, intercept + slope * np.log(x_new))

# Compute r2 again
r2 = r2_score(y_new, y_pred)
print(f"The coefficient of determination or r squared is {r2}")
```

```python
residual = y_new - y_pred

# Residual plot
plt.scatter(x, residual, c='red', label='Residuals')
plt.axhline(y=0, color='blue', linestyle='--', label='Zero Residual Line')
plt.xlabel('X')
plt.ylabel('Residuals')
plt.legend()
plt.title('Residual Plot Using Logarithmic Model')
plt.grid(True)
plt.show()
```
